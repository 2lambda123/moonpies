```python
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir('/home/cjtu/projects/moonpies/moonpies')
import moonpies as mp
import config
cfg = config.Cfg()
rng = mp.get_rng(0)
```

```python
regime = 'd'
time_arr = mp.get_time_array(cfg)
mindiam, maxdiam, step = cfg.diam_range[regime]
sfd_slope = cfg.sfd_slopes[regime]
diams, sfd_prob = mp.get_crater_sfd(mindiam, maxdiam, step, sfd_slope, cfg.dtype)
num_craters_t = mp.num_craters_chronology(mindiam, maxdiam, time_arr, cfg)

# Round to integer number of craters
num_craters_t = mp.probabilistic_round(num_craters_t, rng=rng)

# Random resample diams num_craters times for each timestep
# Note: list much faster than np array here b/c num_craters drops exponentially
random_diams_t = []
for num_craters in num_craters_t:
    random_diams_t.append(np.random.choice(diams, num_craters, p=sfd_prob))

a = np.array(random_diams_t)
```
