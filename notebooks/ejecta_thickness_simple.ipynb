```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Ejecta Module Equations: Simple Craters

def ejecta_thickness_first(d, r):
    """ Return ejecta thickness given (Delta: \u03B4) given simple crater radius (r, meters) and distance from crater rim (d, meters) in the first order: -2.5
    
    Explicitly applies to continuous ejecta (hypothetical 4*crater radii)"""
    
    return 0.14 * r * ((d/r)**(-2.5))

def ejecta_thickness_second(d, r):
    """ Return ejecta thickness (Delta: \u03B4) given simple crater radius (r, meters) and distance from crater rim (d, meters) in the second order: -3.0
    
    Explicitly applies to continuous ejecta (hypothetical 4*crater radii)"""
    
    return 0.14 * r * ((d/r)**(-3.0))

def ejecta_thickness_third(d, r):
    """Return ejecta thickness (Delta: \u03B4) given simple crater radius (r, meters) and distance from crater rim (d, meters) in the third order: -3.5
    
    Explicity applies to continuous ejecta (hypotehtical 4*crater radii)"""
    
    return 0.14 * r * ((d/r)**(-3.5))
```

```python
#Variables for thickness equations
""" Below are calls to incorporate multiple simple craters; the example below only shows ejecta distance for Shackleton Simple Equator"""
#rad = pd.read_csv('cannon2020_crater_ages_python.csv', usecols=[9]) #Pulls Craters Diameter(km) from Cannon
#Crater_Names = pd.read_csv('cannon2020_crater_ages_python.csv', usecols=[0])

""" Thickness for Shackleton Begins Here """

r = (20.9*1000)/2 #Radius of shackleton from cannon, in meters (m)
d = np.arange(650)*1000 #1D Array of distance values from 0 to 650 km --> 0 to 650,000 in meters

thickness_first = np.zeros((1, len(d))) #Empty arrays with (x,y) dimensions of (# of craters of interest(1 for shackleton), 0 to max distance values)
thickness_second = np.zeros((1, len(d))) #Empty arrays with (x,y) dimensions of (# of craters of interest(1 for shackleton), 0 to max distance values)
thickness_third = np.zeros((1, len(d))) #Empty arrays with (x,y) dimensions of (# of craters of interest(1 for shackleton), 0 to max distance values)

#Adding values of thickness to empty arrays 
for i in range(0,1):
    thickness_first[i,:] = ejecta_thickness_first(np.linspace(1.*r, 4.*r, len(d)), r)
"""for value r: start at 1, end at 4*crater radii, do this for the number of values in d"""

for i in range(0,1):
    thickness_second[i,:] = ejecta_thickness_second(np.linspace(1.*r, 4.*r, len(d)), r)
    
for i in range(0,1):
    thickness_third[i,:] = ejecta_thickness_third(np.linspace(1.*r, 4.*r, len(d)), r)

#Lets Plot
f, ax = plt.subplots(3, figsize =(10,15))

#First Order Equation
for i in range(0,1):
    ax[0].semilogy(np.linspace(1.*r,4.*r, len(d))/r, thickness_first[i,:], label='Thickness')
ax[0].set_title('Thickness of Ejecta (First Order)')
ax[0].set_xlabel('Distance (Crater Radii)')
ax[0].set_ylabel('Thickness ($m$)')
ax[0].grid(True, which="both")
#ax[0].legend(thickness_first[i,:], labels='Shackleton Crater', bbox_to_anchor=(1,1))

#Second Order Equation
for i in range(0, 1):
    ax[1].semilogy(np.linspace(1.*r, 4.*r, len(d))/r, thickness_second[i,:], label='Thickness')
ax[1].set_title('Thickness of Ejecta (Second Order)')
ax[1].set_xlabel('Distance (Crater Radii)')
ax[1].set_ylabel('Thickness ($m$)')
ax[1].grid(True, which="both")
#ax[1].legend(thickness_second[i,:], labels='Shackleton Crater')

#Third Order Equation
for i in range(0,1):
    ax[2].semilogy(np.linspace(1.*r, 4.*r,len(d))/r, thickness_third[i,:], label='Thickness')
ax[2].set_title('Thickness of Ejecta (Third Order)')
ax[2].set_xlabel('Distance (Crater Radii)')
ax[2].set_ylabel('Thickness ($m$)')
ax[2].grid(True, which="both")
#ax[2].legend(thickness_third[i,:], labels='Sh')

plt.tight_layout()
```
